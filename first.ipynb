## Null value handling

```python
import numpy as np
arr=np.array([1,7,3,np.nan,np.nan,np.nan,8,6,3,2,3,6,7,7,8,8])
# print(arr)
s=pd.Series(arr)

print(s.isnull().sum())
print(s.notnull().sum())
print(s.hasnans)
```
